```python
# 0) Environment
import sys, os, platform
print("Python:", sys.version)

import tensorflow as tf
print("TensorFlow:", tf.__version__)
print("GPU available:", tf.config.list_physical_devices('GPU'))

# Reduce TF logs
os.environ.setdefault("TF_CPP_MIN_LOG_LEVEL", "2")

# Basics
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# Robust image loading
from PIL import Image, ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
# 1) Setup & Data Loading
# Point to your dataset root that contains 'Healthy' and 'Diseased'
data_dir = Path(r"C:\Users\hp\Desktop\irisdata").resolve()
print("Using data dir:", data_dir)

healthy_dir = data_dir / 'Healthy'
diseased_dir = data_dir / 'Diseased'
assert healthy_dir.exists() and diseased_dir.exists(), "Folders 'Healthy' and 'Diseased' not found."

healthy_files = list(healthy_dir.glob('*'))
diseased_files = list(diseased_dir.glob('*'))
print("Healthy files:", len(healthy_files))
print("Diseased files:", len(diseased_files))
assert len(healthy_files) >= 1 and len(diseased_files) >= 1, "No images found."

# Visualize a few images
import random
sample_paths = random.sample(healthy_files, min(3, len(healthy_files))) + random.sample(diseased_files, min(3, len(diseased_files)))

plt.figure(figsize=(10,6))
for i, p in enumerate(sample_paths):
    ax = plt.subplot(2, 3, i+1)
    img = Image.open(p).convert('RGB')
    plt.imshow(img)
    plt.title("Healthy" if p.parent.name == "Healthy" else "Diseased")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# 2) Preprocessing
# 224x224, normalize [0,1], 80/20 split
from tensorflow import keras
from tensorflow.keras import layers

IMG_SIZE = (224, 224)
BATCH_SIZE = 16
SEED = 1337
CLASS_NAMES = ['Healthy', 'Diseased']  # enforce class order

train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    labels='inferred',
    label_mode='binary',
    validation_split=0.2,
    subset='training',
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    color_mode='rgb',
    class_names=CLASS_NAMES,
    shuffle=True
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    labels='inferred',
    label_mode='binary',
    validation_split=0.2,
    subset='validation',
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    color_mode='rgb',
    class_names=CLASS_NAMES,
    shuffle=False
)

print("Classes:", train_ds.class_names)

AUTOTUNE = tf.data.AUTOTUNE

def preprocess(x, y):
    x = tf.cast(x, tf.float32) / 255.0
    return x, y

data_augmentation = keras.Sequential([
    layers.RandomFlip('horizontal'),
    layers.RandomRotation(0.05),
    layers.RandomZoom(0.05),
])

train_ds = (
    train_ds
    .map(preprocess, num_parallel_calls=AUTOTUNE)
    .map(lambda x, y: (data_augmentation(x, training=True), y), num_parallel_calls=AUTOTUNE)
    .cache().prefetch(AUTOTUNE)
)
test_ds = test_ds.map(preprocess, num_parallel_calls=AUTOTUNE).cache().prefetch(AUTOTUNE)

# Quick visualization post-preprocess
plt.figure(figsize=(8, 8))
for images, labels in train_ds.take(1):
    for i in range(min(9, images.shape[0])):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy())
        lbl = 'Healthy' if int(labels[i].numpy()) == 0 else 'Diseased'
        plt.title(lbl)
        plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# 3) Transfer Learning Model (MobileNetV2)
base_model = tf.keras.applications.MobileNetV2(
    include_top=False,
    weights='imagenet',
    input_shape=IMG_SIZE + (3,)
)
base_model.trainable = False  # freeze for speed

inputs = keras.Input(shape=IMG_SIZE + (3,))
x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.2)(x)
outputs = layers.Dense(1, activation='sigmoid')(x)
model = keras.Model(inputs, outputs)

model.compile(
    optimizer=keras.optimizers.Adam(1e-3),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)
model.summary()
```

```python
# Train for 3 epochs
EPOCHS = 3
history = model.fit(
    train_ds,
    validation_data=test_ds,
    epochs=EPOCHS,
)
```

```python
# 4) Evaluation: accuracy, confusion matrix, curves
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report

y_true, y_pred = [], []
for batch_images, batch_labels in test_ds:
    preds = model.predict(batch_images, verbose=0)
    y_pred.extend((preds.ravel() > 0.5).astype(int))
    y_true.extend(batch_labels.numpy().astype(int).ravel())

y_true = np.array(y_true)
y_pred = np.array(y_pred)
acc = (y_true == y_pred).mean()
print(f"Test Accuracy: {acc:.4f}")

cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:\n", cm)
print("\nClassification Report:\n", classification_report(y_true, y_pred, target_names=CLASS_NAMES))

# Curves
hist = history.history
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(hist['loss'], label='train_loss')
plt.plot(hist['val_loss'], label='val_loss')
plt.title('Loss'); plt.legend()
plt.subplot(1,2,2)
plt.plot(hist['accuracy'], label='train_acc')
plt.plot(hist['val_accuracy'], label='val_acc')
plt.title('Accuracy'); plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 5) Explainability: Grad-CAM overlays
from PIL import Image
import matplotlib.cm as cm
import tensorflow as tf
import numpy as np
from pathlib import Path # Already imported, just for clarity

# Ensure Image.Resampling is defined globally once, or handle within the function
try:
    if not hasattr(Image, 'Resampling'):
        Image.Resampling = Image
except NameError:
    # If Image was not imported correctly, try to import it now
    from PIL import Image
    if not hasattr(Image, 'Resampling'):
        Image.Resampling = Image

# Use the existing base_model (MobileNetV2) and model objects

def get_img_and_feature_extractor(model, target_layer_name):
    """
    Creates a new Keras Model that outputs the target feature maps.
    """
    # The feature extractor goes from base_model's input to the target layer's output
    feature_extractor = tf.keras.models.Model(
        inputs=base_model.input, 
        outputs=base_model.get_layer(target_layer_name).output
    )
    return feature_extractor

def make_gradcam_heatmap(img_array, model, feature_extractor):
    
    # 1. Compute gradients
    with tf.GradientTape() as tape:
        # Cast input array
        input_tensor = tf.cast(img_array, tf.float32)
        
        # *** NEW CRUCIAL FIX: Watch the input tensor ***
        # Watching the input often forces the tracing through the frozen layers
        # when an intermediate tensor is also watched.
        tape.watch(input_tensor) 

        # Get the feature map from the feature extractor model
        feature_maps = feature_extractor(input_tensor)
        
        # Explicitly watch the feature maps for a frozen model (needed alongside watching input)
        tape.watch(feature_maps)
        
        # Get the final prediction from the full model
        predictions = model(input_tensor)
        
        # Use the probability of the predicted class (for the sigmoid output)
        loss = predictions[:, 0]
        
    # Compute gradients of the prediction w.r.t the feature maps
    grads = tape.gradient(loss, feature_maps)
    
    # Check if gradients are None (if still None, we must stop)
    if grads is None:
        print("WARNING: Gradients are still None after all fixes. Grad-CAM will likely show noise.")
        # Return a blank heatmap to prevent crash
        feature_maps_shape = feature_extractor(input_tensor).shape
        return np.zeros(feature_maps_shape[1:3]) 
        
    # Global average pool the gradients (Grad-CAM weights)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    # 2. Create the heatmap
    # Multiply each feature map by its corresponding weight (pooled gradient)
    feature_maps = feature_maps[0] # remove batch dimension
    heatmap = tf.reduce_sum(tf.multiply(pooled_grads, feature_maps), axis=-1)
    
    # Apply ReLU and normalize
    heatmap = tf.maximum(heatmap, 0) / (tf.reduce_max(heatmap) + 1e-8)
    
    return heatmap.numpy()


def overlay_and_save(img, heatmap, alpha=0.4, save_path='overlay.png'):
    # img: float32 [0,1], shape (H,W,3)
    cmap = cm.get_cmap('jet')
    
    # *** FIX for UnboundLocalError: Use the globally imported Image ***
    # Since Image is imported at the top, we rely on that global scope.
    
    # Resize heatmap to match image size using PIL.Image.BILINEAR for smooth upsampling
    h_img = Image.fromarray(np.uint8(255 * heatmap)).resize(
        (img.shape[1], img.shape[0]), Image.Resampling.BILINEAR
    )
    h_arr = np.array(h_img) / 255.0
    
    # Convert heatmap to colored array (H, W, 4) -> (H, W, 3)
    colored = cmap(h_arr)[:, :, :3]
    
    # Combine original image and colored heatmap
    overlay = np.clip(alpha * colored + (1 - alpha) * img, 0, 1)
    
    # Save the resulting image
    Image.fromarray((overlay * 255).astype(np.uint8)).save(save_path)

# --- Main execution block (re-run this part) ---

out_dir = Path('gradcam_output')
out_dir.mkdir(parents=True, exist_ok=True)

TARGET_CONV_LAYER_NAME = 'Conv_1' 

try:
    feature_extractor_model = get_img_and_feature_extractor(model, TARGET_CONV_LAYER_NAME)
except ValueError as e:
    print(f"Error building feature extractor. Check layer name: {e}")
    feature_extractor_model = None

if feature_extractor_model is None:
    print("Could not build feature extractor; skipping Grad-CAM.")
else:
    print(f"Using target layer: {TARGET_CONV_LAYER_NAME}")
    saved = 0
    # Generate overlays for a few validation images
    for batch_images, batch_labels in test_ds.take(10): 
        b = batch_images.numpy()
        for i in range(len(b)):
            if saved >= 3:
                break
                
            img = b[i]
            inp = np.expand_dims(img, axis=0) # Add batch dimension

            # Pass the feature extractor model to the heatmap function
            heatmap = make_gradcam_heatmap(inp, model, feature_extractor_model)
            
            # Get prediction for file naming
            pred = float(model.predict(inp, verbose=0)[0][0])
            pred_class = 'Diseased' if pred > 0.5 else 'Healthy'
            
            # Suppress the DeprecationWarning by ensuring scalar conversion
            true_lbl = 'Healthy' if int(batch_labels[i].numpy().item()) == 0 else 'Diseased'
            
            save_path = out_dir / f'gradcam_{saved+1}_True{true_lbl}_Pred{pred_class}_p{pred:.2f}.png'
            
            overlay_and_save(img, heatmap, save_path=str(save_path))
            saved += 1
            
        if saved >= 3:
            break
            
    print(f"Saved {saved} Grad-CAM images to: {out_dir.resolve()}")
```
